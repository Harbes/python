## _abline_ : Add a Straight Line to a Plot
* __Usage__
```R
  abline(a, b, untf = FALSE, ...) # This form specifies the line in intercept/slope form.(alternatively "a" can be specified on its own and is taken to contain the slope and intercept in vector form).
  abline(h=, untf = FALSE, ...) # The h= and v= forms draw horizontal and vertical lines at the specified coordinates.
  abline(v=, untf = FALSE, ...)
  abline(coef=, untf = FALSE, ...) # "coef" specifies the line by a vector containing the slope and intercept.
  abline(reg=, untf = FALSE, ...) # 'reg' is a regression object which contains "reg$coef". If it is of length 1 then the value is taken to be the slope of a line through the origin, otherwise, the first 2 values are taken to be the intercept and slope.
```

* __Examples__
```R
  data(cars)
  z <- lm(dist ~ speed, data = cars)
  plot(cars)
  abline(z) # 也可以 abline(reg=z[1]) 或 abline(z[1])，但是不能 abline(coef=z[1])  
```

## _arrows_ : Add Arrows to a Plot
* __Usage__
```R
  arrows(x0, y0, x1, y1, length = 0.25, angle = 30, code = 2, # code=2,arrowhead->(x0[i],y0[i]); code=1,arrowhead->(x1[i],y1[i]);code=3,both. Unless length = 0, when no head is drawn.
         col = par("fg"), lty = NULL, lwd = par("lwd"),
         xpd = NULL)
```

* __Examples__
```R
  x <- runif(12); y <- rnorm(12)
  i <- order(x,y); x <- x[i]; y <- y[i]
  plot(x,y, main="arrows(.) and segments(.)")
  ## draw arrows from point to point :
  s <- seq(length(x)-1) # one shorter than data
  arrows(x[s], y[s], x[s+1], y[s+1], col= 1:3)
  s <- s[-length(s)]
  segments(x[s], y[s], x[s+2], y[s+2], col='pink')
```

## _axis_ : Add an Axis to a Plot
```R
  axis(side, # 1=below, 2=left, 3=above and 4=right. 逆时针

       tick = TRUE, # a logical value specifying whether tickmarks should be drawn
       line = NA, pos = NA, outer = FALSE, font = NA, 
       vfont = NULL, lty = "solid", lwd = 1, col = NULL, ...)
```

## _assocplot_ : Association Plots
  Produce a Cohen-Friendly association plot indicating deviations from independence of rows and columns in a 2-dimensional contingency table.
* __Usage__
```R
  assocplot(x, col = c("black", "red"), space = 0.3,
            main = NULL, xlab = NULL, ylab = NULL)
```

* __Examples__
```R
  data(HairEyeColor)
  ## Aggregate over sex:
  x <- margin.table(HairEyeColor, c(1, 2))
  x
  assocplot(x, main = "Relation between hair and eye color")
```

````r
data(HairEyeColor)
## Aggregate over sex:
x <- margin.table(HairEyeColor, c(1, 2))
x
assocplot(x, main = "Relation between hair and eye color")
````

## _bandwidth_ : Bandwidth selectors for gaussian windows in _density_
* __Usage__
```R
bw.nrd0(x)
  bw.nrd(x)
  bw.ucv(x, nb = 1000, lower, upper)
  bw.bcv(x, nb = 1000, lower, upper)
  bw.SJ(x, nb = 1000, lower, upper, method = c("ste", "dpi"))
```
* __Examples__
```R
  options(warn=-1)
  data(precip)
  plot(density(precip, n = 1000))
  rug(precip)
  lines(density(precip, bw="nrd"), col = 2)
  lines(density(precip, bw="ucv"), col = 3)
  lines(density(precip, bw="bcv"), col = 4)
  lines(density(precip, bw="SJ-ste"), col = 5)
  lines(density(precip, bw="SJ-dpi"), col = 6)
  legend(55, 0.035,
    legend = c("nrd0","nrd","ucv","bcv","SJ-ste","SJ-dpi"),
    col = 1:6, lty = 1) # col=rainbow(6)
```

## Beta
* __Usage__

```R
 dbeta(x,shape1,shape2,ncp=0,log=FALSE) # density
 pbeta(q,shape1,shape2,ncp=0,lower.tail=TRUE,log.p=FALSE) # distribution function
 qbeta(p,shape1,shape2,lower.tail=TRUE,log.p=FALSE) # quantile function 
 rbeta(n,shape1,shape2) # random deviates
```

* __Details__
  
  The Beta distribution with parameters ___shape1 = a___ and ___shape2 = b___ has density $$ f(x)=\frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}x^a(1-x)^b $$ for $a > 0, b > 0$ and $0 ≤ x ≤ 1$ where the boundary values at $x = 0$ or
$x = 1$ are defined as by continuity (as limits).

* __Examples__
```R
  x <- seq(0, 1, length=21)
  d <- dbeta(x, 3, 2)
  p <- pbeta(x, 3, 2)
  plot(x,d)
  plot(x,p)
```

````r
x <- seq(0, 1, length=21)
d <- dbeta(x, 3, 2)
p <- pbeta(x, 3, 2)
plot(x,d)
lines(x,p)
````
